```python
# ── Cell 2: City name → latitude/longitude ───────────────────────────
# Uses Open-Meteo's free geocoding API — no key needed

def get_city_coordinates(city_name):
    """
    Takes a city name string, returns (lat, lon, display_name).
    Raises ValueError if city is not found.
    """
    url = "https://geocoding-api.open-meteo.com/v1/search"
    params = {
        "name": city_name,
        "count": 1,
        "language": "en",
        "format": "json"
    }

    response = requests.get(url, params=params)
    response.raise_for_status()
    data = response.json()

    if "results" not in data or len(data["results"]) == 0:
        raise ValueError(f"City '{city_name}' not found. Try a larger nearby city.")

    result = data["results"][0]
    lat = result["latitude"]
    lon = result["longitude"]
    display_name = f"{result['name']}, {result.get('country', '')}"

    return lat, lon, display_name


# ── Interactive test — you supply the city ───────────────────────────
city_input = input("Enter a city name: ")
lat, lon, name = get_city_coordinates(city_input)
print(f"✓ Found: {name}")
print(f"  Latitude : {lat}")
print(f"  Longitude: {lon}")
```

```python
# ── Cell 3: Fetch historical weather data from Open-Meteo ────────────
# Uses lat/lon from Cell 2 — run Cell 2 first

def fetch_weather_data(lat, lon, start_year=2020, end_year=2024):
    """
    Fetches daily historical weather for a lat/lon coordinate.
    Returns raw JSON response from Open-Meteo.
    start_year / end_year: range of historical data to pull.
    """
    url = "https://archive-api.open-meteo.com/v1/archive"
    params = {
        "latitude":  lat,
        "longitude": lon,
        "start_date": f"{start_year}-01-01",
        "end_date":   f"{end_year}-12-31",
        "daily": [
            "temperature_2m_max",
            "temperature_2m_min",
            "precipitation_sum",
            "windspeed_10m_max"
        ],
        "timezone": "auto"
    }

    response = requests.get(url, params=params)
    response.raise_for_status()
    raw = response.json()

    # Basic sanity check
    if "daily" not in raw:
        raise ValueError("Unexpected API response — 'daily' key missing.")

    print(f"✓ Fetched data for coordinates ({lat}, {lon})")
    print(f"  Date range : {params['start_date']} → {params['end_date']}")
    print(f"  Records    : {len(raw['daily']['time'])} days")

    return raw


# ── Run using city from Cell 2 ───────────────────────────────────────
raw_data = fetch_weather_data(lat, lon)
```

temp_range

1.   Derived column: max minus min temp per day. Unusually large swings are strong anomaly signals — Person 2's model will pick these up.
2. fillna(median)
Median is safer than mean for weather data — extreme values (the anomalies we're hunting) would skew the mean and corrupt the fill.
3. set_index("date")
Makes date the row label instead of a column — standard practice for time-series DataFrames and makes Plotly charts cleaner.

```python
# ── Cell 4: Parse raw JSON into a clean Pandas DataFrame ─────────────
# Uses raw_data from Cell 3 — run Cell 3 first

def parse_weather_dataframe(raw, city_name):
    """
    Converts raw Open-Meteo JSON into a clean DataFrame.
    Renames columns, sets date as index, fills missing values.
    """
    daily = raw["daily"]

    df = pd.DataFrame({
        "date":          pd.to_datetime(daily["time"]),
        "temp_max":       daily["temperature_2m_max"],
        "temp_min":       daily["temperature_2m_min"],
        "precipitation":  daily["precipitation_sum"],
        "windspeed":      daily["windspeed_10m_max"],
    })

    # Set date as index — makes time-series operations easier
    df = df.set_index("date")

    # Add a derived column — daily temp range
    df["temp_range"] = df["temp_max"] - df["temp_min"]

    # Fill any missing values with column median (safer than dropping)
    df = df.fillna(df.median(numeric_only=True))

    # Tag the city for reference
    df["city"] = city_name

    print(f"✓ DataFrame ready for: {city_name}")
    print(f"  Shape   : {df.shape[0]} rows × {df.shape[1]} columns")
    print(f"  Columns : {list(df.columns)}")
    print(f"  Nulls   : {df.isnull().sum().sum()} remaining")
    print()

    return df


# ── Run using raw_data + city name from Cell 2 ───────────────────────
df = parse_weather_dataframe(raw_data, name)

# Preview first 5 rows
df.head()
```
